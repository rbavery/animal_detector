```python
import torch
import torch.utils.data as data
from torch.utils.data import DataLoader, SubsetRandomSampler
from torchvision.datasets import ImageFolder
from torchvision import transforms
```

```python
from pathlib import Path
cwd = Path.cwd()
data_dir = cwd/'data'/'train'
!ls {data_dir}
```

```python
transform1 = transforms.Compose([
    transforms.Resize(int(224*1.2)),
    transforms.RandomResizedCrop(224),
#     transforms.RandomResizedCrop(224),
    transforms.ToTensor()
])

transform2 = transforms.Compose([
    transforms.ToTensor()
])

ds1 = ImageFolder(root=data_dir, transform=transform1)
ds2 = ImageFolder(root=data_dir, transform=transform2)

dl1 = DataLoader(ds1, batch_size=1, shuffle=False, sampler=SubsetRandomSampler([2]))
dl2 = DataLoader(ds2, batch_size=1, shuffle=False, sampler=SubsetRandomSampler([2]))
```

```python
img1, label1 = next(iter(dl1))
img2, label2 = next(iter(dl2))
```
